Silahkan coba berikut ini:

```python
a=10 #inilisiasi a adalah 10
b=5 #inilisiasi b adalah 5
selisih = a-b #pengurangan
jumlah = a+b #pertambahan
kali = a*b #perkalian
bagi = a/b #pembagian
print("Hasil penjumlahan dan b adalah", jumlah) #menampilkan penjumlahan
print("Selisih a dan b adalah :", selisih) #menampilkan pengurangan
print("Hasil perkalian a dan b adalah :", kali) #menampilkan perkalian
print("Hasil pembagian a dan b adalah:", bagi) #menampilkan pembagian
```
