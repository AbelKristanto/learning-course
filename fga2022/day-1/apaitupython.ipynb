# Apa itu python?

Python merupakan bahasa pemrograman yang diciptakan pada tahun 1991 oleh Guido van Russom, seorang matematikawan. Python dapat dikatakan sebagai programming language dengan perkembangan tercepat. Perkembangan ini didukung oleh library atau pustaka python yang cukup melimpah. Kegunaan python pun sangatlah luas diantaranya:
1. Data Science
2. Web-development
3. Game-development
4. Computer Vision
5. dan sebagainya

Python merupakan salah satu programming language yang cocok untuk scripting dan bisa berjalan dalam berbagai platform OS dan IoT yang mana tergolong ke dalam high level programming language, python sangat mudah dimengerti karena syntax-nya yang sederhana. 

Adapun kelebihan dengan menggunakan python diantaranya:
1. Python dirancang untuk mudah ditulis dan dibaca, serta memiliki beberapa kesamaan dengan bahasa Inggris dengan pengaruh dari matematika.
2. Python menggunakan baris baru untuk mengakhiri perintah, dibandingkan dengan bahasa pemrograman lain yang sering menggunakan titik koma atau tanda kurung.
3. Python bergantung pada indentasi, menggunakan spasi, untuk mendefinisikan ruang lingkup; seperti lingkup loop, fungsi, dan kelas. Bahasa pemrograman lainnya sering menggunakan kurung kurawal untuk tujuan ini
4. Jika kalian mencari di internet tentang programming python, biasanya akan ada 2 jenis python, python 2 dan python 3. Versi utama Python yang paling baru adalah Python 3. Saat ini, python 2 sudah memasuki end-of-life pada Januari 2020.

Penggunaan variabel atau suatu objek yang bisa merepresentasikan sebuah nilai atau value sangat penting dalam bahasa pemrograman. Selain itu mempermudah dalam membaca source code, pemberian variable yang efisien juga akan membuat code berjalan optimal dan dinamis. Variabel dalam python terbagi diantaranya sebagai berikut ini:
    


```python
# Silahkan coba untuk printing berikut ini:

# inputan data
var_string="Belajar Python yuk"
var_int=10
var_float=3.14
var_list=[1,2,3,4]
var_tuple=("satu","dua","lima")
var_dict={"nama":"Budi", 'umur':30}

# memanggil data yang ditulis diatas
print(var_string)
print(var_int)
print(var_list)
print(var_tuple)
print(var_dict)

# mengetahui type data yang digunakan
print(type(var_string))
print(type(var_int))
print(type(var_float))
print(type(var_tuple))
print(type(var_dict))
```
