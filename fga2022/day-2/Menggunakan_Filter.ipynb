<h1>Menggunakan Filter</h1>

Dalam SQL ada cara mudah untuk melakukan filter yakni klausul WHERE:
1. Filter data dengan kondisi teks tertentu.
2. Filter data dengan nilai angka tertentu.
3. Filter data dengan dua kondisi menggunakan operator AND dan OR.

Sebagai contoh penulisan sebagai berikut ini:
```sql
SELECT column_name FROM table_name WHERE condition
```

````python
sql(
"""
SELECT * FROM orders 
WHERE ship_country = 'Germany';
""")
````

Selanjutnya kita akan menggunakan operand OR
```sql
SELECT column_name FROM table_name WHERE condition1 OR condition2
```

OR artinya mengeluarkan hasil sesuai condition1 jika tidak memenuhi condition2 [SALAH SATU TERPENUHI].

````python
sql(
"""
SELECT * FROM orders 
WHERE ship_country = 'Germany'
OR ship_country = 'France';
""")
````

Selanjutnya apakah filter bisa untuk angka? Tentu bisa. Tapi tanpa kutip ya! karena bukan string.
```sql
SELECT column_name FROM table_name WHERE condition1 > 50
```

````python
sql(
"""
SELECT * FROM products
WHERE units_in_stock > 100;
""")
````

Selanjutnya kita akan menggunakan operand AND

```sql
SELECT column_name FROM table_name WHERE condition1 AND condition2
```

AND artinya mengeluarkan hasil yang memenuhi dua kondisi ya!

````python
sql(
"""
SELECT * FROM products
WHERE units_in_stock > 100 AND
product_name = 'Boston Crab Meat';
""")
````

**Kesimpulan**
1. Filter di SQL diimplementasikan dengan menggunakan WHERE diikuti dengan satu atau lebih kondisi logis.
2. Kondisi logis ini memiliki format nama kolom diikuti dengan nilai yang akan difilter. Untuk teks sederhana bisa menggunakan tanda sama dengan, sedangkan untuk nilai angka bisa menggunakan operator perbandingan matematika.
