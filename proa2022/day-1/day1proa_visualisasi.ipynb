````r
# Ketik disini
library(ggplot2)
````

````r
str(mtcars)
````

````r
# Cara sederhana membuat grafik
ggplot(mtcars, aes(cyl, mpg)) +
  geom_point() # Geometry
````

````r
# apabila cyl diubah menjadi faktor
# ini akan menghilangkan 5, 7 karena dibuat menjadi kategorikal bukan value
ggplot(mtcars, aes(factor(cyl), mpg)) +
  geom_point()
````

````r
# Mengatur warna berubah seiring dengan ukuran disp
ggplot(mtcars, aes(wt, mpg, color = disp)) +
  geom_point()
````

````r
# Warna berubah dengan ukuran yang berbeda
ggplot(mtcars, aes(wt, mpg, color = disp, size=disp)) +
  geom_point()
````

````r
# Mengatur dengan perbedaan shape (perlu data kategorikal)
ggplot(mtcars, aes(wt, mpg, shape = factor(cyl))) +
  geom_point()
````

````r
# Memahami geom_jitter untuk mengatasi overplotting dengan penambahan noise acak
# Bandingkan dengan sesudahnya
g <- ggplot(iris, aes(x=Sepal.Length, y=Sepal.Width)) + geom_jitter()
g
````
