<h2>Structured Query Language</h2>
<p>Di bagian ini kita akan membahas tentang bahasa SQL!</p>

Apa itu SQL?

SQL merupakan bahasa pemograman untuk pengolahan database. Bahasa komputer standar yang digunakan untuk berinteraksi dengan suatu sistem database - atau lebih tepatnya sistem manajemen database relasional. Jadi, user dapat menambahkan, mengubah, mengupdate, mencari dan menghapus data dari suatu sistem database dengan menggunakan SQL.

Ada dua jenis kategori dari interaksi SQL:
1. Data Definition Language (DDL): Befokus membuat (CREATE), mengubah (ALTER), menghapus(DROP) struktur penyimpanan data
2. Data Manipulation Language (DML): Berfokus menyisipkan (INSERT), mengambil data (SELECT), mengubah isi data (UPDATE), menghapus data (DELETE).

Contoh penerapan sebagai berikut ini:

```sql
SELECT name_column FROM name_table
```

Lalu, alasan kita mempelajari SQL?
1. Biasanya perusahaan besar menyimpan data mereka dalam database, karena relatif besar datanya
2. Sumber dalam pengambilan keputusan perusahaan
3. Meningkatkan kinerja perusahaan dengan informasi yang tepat

Biasanya SQL mengakses data dari Database Management System (DBMS) yang berisikan semua informasi ataupun analisa yang dibutuhkan manajemen. DBMS berisikan data trafik, penjualan, data produk, identitas konsumen dan sebagainya.

> SQL hanya digunakan di sistem database berbasis Relational Database Management System.

**KESIMPULAN**

SQL singkatan dari Structured Query Language, adalah sebuah bahasa komputer sederhana yang menjadi standar untuk memungkinkan seseorang berkomunikasi dengan suatu sistem database manajemen relasional (RDBMS).

<h2>Sistem Database Relasional</h2>
<p>Di bagian ini kita akan membahas tentang RDBMS itu sendiri</p>

RDBMS berarti **R**elational **D**atabase **M**anagement **S**ystem untuk Membuat, Memperbarui, dan Mengelola suatu basis data relasional (Relational Database). SQL berguna untuk mengakses data tersebut

Basis data relasional sendiri merupakan suatu jenis database dimana data – data umumnya disimpan dalam bentuk yang terstruktur berupa tabel (baris dan kolom) dan setiap tabel/ data yang terdapat dalam database memiliki relasi (relational) satu sama lain. Hal ini menunjukan bahwa data dapat dikelola, scalable, flexible dalam melakukan analisa.

Contoh berikut ini:



Ini artinya bahwa:
1. Basis data rasional cukup mudah dikelola. Setiap tabel/data dapat diupdate atau dimodifikasi tanpa mengganggu tabel/data yang lain.
2. Flexible : jika perlu memperbarui data, hanya perlu melakukannya sekali saja - jadi tidak perlu lagi mengubah banyak file satu per satu. Selain itu, basis data rasional juga cukup mudah untuk di-extend. Misalnya saat data sudah semakin banyak, dapat dengan mudah memperbesar kapasitas dari database yang dimiliki.

Produk RDBMS diantaranya:

<h2>Menggunakan Filter</h2>

Dalam SQL ada cara mudah untuk melakukan filter yakni klausul WHERE:
1. Filter data dengan kondisi teks tertentu.
2. Filter data dengan nilai angka tertentu.
3. Filter data dengan dua kondisi menggunakan operator AND dan OR.

Sebagai contoh penulisan sebagai berikut ini:
```sql
SELECT column_name FROM table_name WHERE condition
```

````python
sql(
"""
SELECT * FROM orders 
WHERE ship_country = 'Germany';
""")
````

Selanjutnya kita akan menggunakan operand OR
```sql
SELECT column_name FROM table_name WHERE condition1 OR condition2
```

OR artinya mengeluarkan hasil sesuai condition1 jika tidak memenuhi condition2 [SALAH SATU TERPENUHI].

````python
sql(
"""
SELECT * FROM orders 
WHERE ship_country = 'Germany'
OR ship_country = 'France';
""")
````

Selanjutnya apakah filter bisa untuk angka? Tentu bisa. Tapi tanpa kutip ya! karena bukan string.
```sql
SELECT column_name FROM table_name WHERE condition1 > 50
```

````python
sql(
"""
SELECT * FROM products
WHERE units_in_stock > 100;
""")
````

Selanjutnya kita akan menggunakan operand AND

```sql
SELECT column_name FROM table_name WHERE condition1 AND condition2
```

AND artinya mengeluarkan hasil yang memenuhi dua kondisi ya!

````python
sql(
"""
SELECT * FROM products
WHERE units_in_stock > 100 AND
product_name = 'Boston Crab Meat';
""")
````

**Kesimpulan**
1. Filter di SQL diimplementasikan dengan menggunakan WHERE diikuti dengan satu atau lebih kondisi logis.
2. Kondisi logis ini memiliki format nama kolom diikuti dengan nilai yang akan difilter. Untuk teks sederhana bisa menggunakan tanda sama dengan, sedangkan untuk nilai angka bisa menggunakan operator perbandingan matematika.

Berikut ini merupakan penggunaan fungsi skalar:

Selanjutnya, kita akan mempelajari bagaimana menggunakan SQRT (Akar kuadrat), seperti berikut ini:

```sql
SELECT SQRT(column_name)
```

## Fungsi Text di SQL

Disini kita akan mempelajari fungsi dalam text yang mana dapat dituliskan sebagai berikut:



Selanjutnya, kita akan mempelajari bagaimana menggunakan CONCAT (Penggabungan dua kolom jadi satu), seperti berikut ini:

```sql
SELECT CONCAT(column_name1, column_name2)
FROM table_name;
```

````python
sql(
"""
SELECT contact_name, CONCAT(address, city) full_address
FROM customers LIMIT 5
"""
)
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan SUBSTRING_INDEX (pemisahan tertentu), seperti berikut ini:

```sql
SELECT SUBSTRING_INDEX(column, delimiter, index to return)
FROM table_name;
```

Ketentuan
1. column --> merupakan nama kolom yang akan dipecah text-nya, 
2. delimiter --> karakter atau gabungan beberapa karakter untuk pemecah text pada kolom bersangkutan, 
3. index_to_return --> indeks dari pecahan text yang akan diambil.

Contoh:
```sql
SELECT SUBSTRING_INDEX("www.genpep.com", ".", 1);
```

Coba [disini](https://www.w3schools.com/sqL/trymysql.asp?filename=trysql_func_mysql_substring_index2)

Selanjutnya, kita akan mempelajari bagaimana menggunakan SUBSTR (pengambilan tertentu), seperti berikut ini:

```sql
SELECT SUBSTR(columnName, Start Index, Number of string to be extract)
FROM table_name;
```

Ketentuan
1. columnName --> nama kolom yang akan dicari substring-nya 
2. Start Index --> indeks dari text yang dimiliki (dimulai dari 1) 
3. Number of string to be extract --> jumlah karakter atau beberapa karakter yang akan diambil.


````python
sql(
"""
SELECT contact_name, SUBSTR(contact_name, 1, 3) call_name
FROM customers LIMIT 5
"""
)
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan LENGTH (menghitung), seperti berikut ini:

```sql
SELECT LENGTH(columnName)
FROM table_name;
```


````python
sql(
"""
SELECT contact_name, SUBSTR(contact_name, 1, 3) call_name, LENGTH(contact_name)
FROM customers LIMIT 5
"""
)
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan REPLACE (mengganti), seperti berikut ini:

```sql
SELECT REPLACE(ColumnName, Character/String to be change, New String/Character)
FROM table_name;
```

Catatan:
1. ColumnName --> nama kolom yang akan diganti isi tiap record/barisnya berdasarkan string/karakter tertentu 
2. Character/String to be change --> string/karakter yang dimiliki untuk diganti 
3. New String/Character --> string/karakter baru pengganti string/karakter sebelumnya

````python
sql(
"""
SELECT contact_name, REPLACE(contact_name, 'Anders', 'Tru')
FROM customers LIMIT 5
"""
)
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan UPPER/LOWER (), seperti berikut ini:

```sql
SELECT UPPER(ColumnName), LOWER(ColumnName)
FROM table_name;
```

````python
sql(
"""
SELECT contact_name, UPPER(contact_name), LOWER(contact_name)
FROM customers LIMIT 5
"""
)
````

## Fungsi Aggregate dan Group By

Bagian sub-bab ini yang mana berisikan berikut ini:


Selanjutnya, kita akan mempelajari bagaimana menggunakan SUM (Menjumlahkan), seperti berikut ini:

```sql
SELECT SUM(ColumnName)
FROM table_name;
```

````python
sql(
"""
SELECT ROUND(SUM(unit_price*quantity))
FROM order_details
"""
)
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan COUNT (Menghitung), seperti berikut ini:

```sql
SELECT COUNT(ColumnName)
FROM table_name;
```

````python
sql(
"""
SELECT COUNT(contact_name)
FROM customers
"""
)
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan AVG (Rata-rata), seperti berikut ini:

```sql
SELECT AVG(ColumnName)
FROM table_name;
```

````python
sql(
"""
SELECT ROUND(AVG(unit_price*quantity))
FROM order_details
"""
)
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan MIN (Nilai terkecil) atau MAX(nilai terbesar), seperti berikut ini:

```sql
SELECT MIN(ColumnName), MAX(ColumnName)
FROM table_name;
```

````python
sql(
"""
SELECT MIN(unit_price), MAX(unit_price)
FROM order_details
"""
)
````

Selanjutnya, kita akan mempelajari dan mengenal bagaimana menggunakan GROUP BY, seperti berikut ini:

```sql
SELECT ColumnName
FROM table_name
WHERE condition
GROUP BY ColumnName
ORDER BY ColumnName;
```

Catatan:
1. GROUP BY digunakan dengan SELECT, artinya kolom yang digunakan di GROUP BY statement, juga perlu ditempatkan di SELECT.
2. GROUP BY ditempatkan setelah WHERE, tetapi jika tidak menggunakan WHERE maka langsung ditempatkan setelah FROM. 
3. Jika menggunakan ORDER BY, maka GROUP BY ditempatkan sebelum ORDER BY. 
4. Group by Single Column, data dikelompokkan menggunakan kriteria dari satu kolom saja, misalnya mengelompokkan data berdasarkan kota saja. 
5. Group by Multiple Column, data dikelompokkan menggunakan kriteria dari dua kolom atau lebih, misalnya mengelompokkan data berdasarkan kota dan merek.

````python
# Single Column
sql(
"""
SELECT ship_country, COUNT(DISTINCT order_id) as total_order, AVG(freight) as avg_price
FROM orders
GROUP BY ship_country
"""
)
````

````python
# Double Column
sql(
"""
SELECT ship_country, ship_name, COUNT(DISTINCT order_id) as total_order, AVG(freight) as avg_price
FROM orders
GROUP BY ship_country, ship_name
"""
)
````

Selanjutnya, kita akan mempelajari dan mengenal bagaimana menggunakan CASE WHEN, seperti berikut ini:

```sql
SELECT ColumnName1, ColumnName2,  
CASE  
    WHEN condition1 THEN result1  
    WHEN condition2 THEN result2  
    WHEN conditionN THEN resultN  
    ELSE result  
END as alias  
FROM TableName;
```

Catatan:

CASE-statement akan mengevaluasi kondisi yang sudah ditentukan, dimulai dari condition1, dan akan mengembalikan hasil (result1), jika condition1 terpenuhi (TRUE). Jika tidak, maka condition2 akan dievaluasi, dan akan mengembalikan result2 jika condition2 terpenuhi, dst. Apabila tidak ada kondisi yang terpenuhi, maka result pada bagian ELSE yang akan dikembalikan. 

````python
# Double Column
sql(
"""
SELECT ship_country, 
COUNT(DISTINCT order_id) as total_order, 
AVG(freight) as avg_price,
CASE
WHEN AVG(freight) > 100 THEN 'Expensive'
WHEN AVG(freight) < 50 THEN 'Cheaper'
ELSE 'Average'
END as remarks
FROM orders
GROUP BY ship_country
"""
)
````

Lalu bagaimana jika menggunakan materi sebelumnya dengan prefix?

````python
sql(
"""
SELECT orders.*, order_details.*  FROM orders, order_details
WHERE orders.order_id=order_details.order_id
LIMIT 5;
""")

# Latihan hubungkan products dengan order_details
````

Lalu bagaimana apabila where dihapuskan? Ini akan menjadi cross join, artinya setiap baris data pada kedua tabel akan dihubungkan satu sama lain - tanpa ada hubungan. Atau sederhananya row_baris_1 * row_baris_2 akan menjadi semua hasil row.

````python
sql(
"""
SELECT orders.*, order_details.*  FROM orders, order_details
LIMIT 5;
""")

# Latihan hubungkan products dengan order_details
````

## INNER JOIN

Penggabungan yang berdasarkan column key yang sama, artinya baris dari kedua atau lebih tabel akan dibandingkan untuk mengecek baris - baris mana saja yang cocok satu sama lain berdasarkan kondisi JOIN yang ditentukan.





Inner Join merupakan irisan kedua tabel_1 dan tabel_2 ya!

````python
sql(
"""
SELECT orders.*, order_details.*  FROM orders
INNER JOIN order_details
ON orders.order_id=order_details.order_id
LIMIT 5;
""")

# Latihan hubungkan products dengan order_details
````

INNER JOIN akan mencocokkan key kolom tabel_1 dengan key kolom lawannya, jika value dari key kolom sama - sama ada di kedua tabel , maka baris itu akan dikembalikan sebagai hasil query dan membentuk satu tabel yang berisi seluruh kolom dari kedua tabel, sedangkan jika value key kolom hanya terdapat di satu kolom, maka baris ini tidak akan ditampilkan.

> Perlu diperhatikan jika menampilkan kolom dengan nama yang sama di kedua tabel, maka pada bagian Select, tidak bisa hanya mengetikkan nama kolom saja, tetapi juga harus didahului oleh prefix nama tabel dimana kolom itu berasal untuk menghindari error karena ambiguitas.

## UNION

Ini digunakan untuk mengabungkan seluruh row tanpa mempedulikan key_column, dengan catatan:
1. Setiap hasil dari SELECT statement yang akan digabungkan (UNION) memiliki jumlah kolom yang sama (Conforming). Ini artinya perlu menyelaraskan urutan kolom yang dimiliki ya!
2. Kolom tersebut juga harus memiliki tipe data yang sama, dan
3. Kolom tersebut memiliki urutan posisi yang sama.

````python
sql(
"""
SELECT * FROM order_details
UNION 
SELECT * FROM order_details
LIMIT 5;
""")

# Latihan hubungkan products
````

````python
sql(
"""
SELECT * FROM order_details
WHERE order_id = '10805'
UNION 
SELECT * FROM order_details
WHERE order_id = '10805'
LIMIT 5;
""")

# Latihan hubungkan products
````

**Perbedaan Union dan JOIN**

Pada metode JOIN, penggabungan dilakukan berdasarkan key/kolom tertentu yang terdapat di tabel-tabel yang akan digabungkan dan key/kolom ini memiliki nilai yang saling terkait. 


Untuk UNION seperti yang sudah dijelaskan, digunakan ketika ingin menggabungkan tabel secara secara vertikal yaitu menggabungkan baris/row dari dua atau lebih tabel. Tidak seperti JOIN, untuk penggabungan dengan UNION, tidak diperlukan key/kolom yang saling terkait tetapi UNION mensyaratkan bahwa jumlah kolom dari tabel - tabel yang akan digabungkankan adalah sama dan berada diposisi yang sama pula.


**KESIMPULAN**

1. Setiap hasil dari SELECT statement yang akan digabungkan (UNION) memiliki jumlah kolom yang sama
2. Kolom tersebut juga harus memiliki tipe data yang sama, dan
3. Kolom tersebut memiliki urutan posisi yang sama.

> Perbedaan mendasar dari JOIN dan UNION adalah JOIN menggabungkan 2 tabel atau lebih berdasarkan baris yang saling berelasi/terkait sedangkan UNION menggabungkan 2 tabel secara vertikal.
