<h1>Fungsi di SQL</h1>
<p>Di sini kita akan mempelajari beberapa fungsi di SQL yang melakukan operasi data di database.</p>

SQL membagi atas dua tipe fungsi:
1. Operasi ini bisa berupa kalkulasi numerik seperti sum, count, avg, etc (SCALAR)  digunakan untuk mengembalikan nilai tunggal (single value) dari suatu nilai input yang diberikan
2. Operasi non-numerik seperti string concatenations dan sub-strings (AGGREGATE)

Berikut ini merupakan penggunaan fungsi skalar:

![image.png](https://drive.google.com/uc?export=view&id=1CcjClDtWXjL7hVszll82hn3ouNWmnc9D)

Pertama, kita akan mempelajari bagaimana menggunakan ABS, seperti berikut ini:

```sql
SELECT ABS(column_name)
FROM table_name;
```

````python
sql(
"""
SELECT product_name, unit_price AS before, ABS(unit_price) AS after
FROM products;
""")
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan CEILING (pembulatan keatas), seperti berikut ini:

```sql
SELECT CEILING(column_name)
FROM table_name;
```

````python
sql(
"""
SELECT product_name, unit_price AS before, CEILING(unit_price) AS after
FROM products;
""")
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan FLOOR (pembulatan kebawah), seperti berikut ini:

```sql
SELECT FLOOR(column_name)
FROM table_name;
```

````python
sql(
"""
SELECT product_name, unit_price AS before, FLOOR(unit_price) AS after
FROM products;
""")
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan ROUND (pembulatan), seperti berikut ini:

```sql
SELECT ROUND(column_name)
FROM table_name;
```

````python
sql(
"""
SELECT ROUND(235.415, 2) AS RoundValue;
"""
)
````

````python
sql(
"""
SELECT product_name, unit_price AS before, ROUND(unit_price) AS after
FROM products;
""")
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan SQRT (Akar kuadrat), seperti berikut ini:

```sql
SELECT SQRT(column_name)
FROM table_name;
```

````python
sql(
"""
SELECT product_name, unit_price AS before, SQRT(unit_price) AS after
FROM products;
""")
````

Selanjutnya, kita akan mempelajari bagaimana menggunakan MOD (Nilai sisa pembagian), seperti berikut ini:

```sql
SELECT MOD(column_name, 2)
FROM table_name;
```

````python
sql(
"""
SELECT MOD(18, 4);
""")
````
